```python
import pandas as pd
import numpy as np
```

```python
Dataset = pd.read_csv("heart disease.csv")
```

```python
Dataset.info()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
data_nor = scale(Dataset)
```

```python
data_nor
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(data_nor)
```

```python
pca_values
pca_data = pd.DataFrame(pca_values)
pca_data
```

```python
var = pca.explained_variance_ratio_
var
```

```python
#cumulative var

var1 = np.cumsum(np.round(var, decimals =2)*100)
var1
```
