```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
Dataset = pd.read_csv("wine.csv")
Dataset.info()
```

```python
data_nor = scale(Dataset)
```

```python
data_nor
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(data_nor)
pca_data = pd.DataFrame(pca_values)
pca_data
```

```python
# variance

var = pca.explained_variance_ratio_
var
```

```python
#cumulative variance
var1 = np.cumsum(np.round(var, decimals=4)*100)
var1
```

```python
final = pca_data.iloc[:,[0,1,2]]
final
```
