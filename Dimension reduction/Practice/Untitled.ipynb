```python
import pandas as pd
import numpy as np
```

```python
Dataset = pd.read_excel("University_Clustering.xlsx")
```

```python
Dataset.describe()
```

```python
Dataset.info()
```

```python
# PCA is only for numeric data, so remove  non-numeric data

data = Dataset.drop(['State'], axis = 1)
```

```python
data.info()
```

```python
from sklearn.decomposition import PCA
# import PCA packages
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

#scale func() is used for standardization
```

```python
# lets have only numeric data in dataset
num_data = data.iloc[:,1:]
```

```python
data_nor = scale(num_data)
```

```python
data_nor
```

```python
pca = PCA(n_components = 6)
# original dataset consists of 6 colums, so components is 6
pca_values = pca.fit_transform(data_nor)
```

```python
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
pca.components_
```

```python
pca.components_[0]
```

```python
# cumulative variance 

var1 = np.cumsum(np.round(var, decimals=4)*100)
var1
```

```python
pca_values

# make the pca_values into a dataframe

pca_data = pd.DataFrame(pca_values)
pca_data
```

```python
# add columns
pca_data.columns = "comp0","comp1","comp2","comp3","comp4","comp5"
pca_data
```

```python
final = pd.concat([Dataset.Univ, pca_data.iloc[:,:3]], axis = 1)
```
