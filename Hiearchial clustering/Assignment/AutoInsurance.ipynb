```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Insur1 = pd.read_csv("AutoInsurance.csv")
Insur1.columns
```

```python
Insur1.info()
```

```python
from feature_engine.outliers import Winsorizer
```

```python
winsor = Winsorizer(capping_method="iqr", tail = 'both', fold =1.5, variables=['Customer Lifetime Value','Income', 'Monthly Premium Auto',
       'Months Since Last Claim', 'Months Since Policy Inception',
       'Number of Open Complaints', 'Number of Policies','Total Claim Amount'])
```

```python
Insur_prepro = winsor.fit_transform(Insur1[['Customer Lifetime Value','Income', 'Monthly Premium Auto',
       'Months Since Last Claim', 'Months Since Policy Inception',
       'Number of Open Complaints', 'Number of Policies','Total Claim Amount']])
Insur_prepro

# Preprocessing done
```

```python
Insur_prepro.describe()
```

```python
# normalization

def norm(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
Insur_norm = norm(Insur_prepro)
Insur_norm = Insur_norm.drop(['Number of Open Complaints'], axis=1)
Insur_norm.info()
```

```python
Insur_norm.describe()
```

```python
#create the dendrogram

from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
```

```python
z = linkage(Insur_norm, method = 'ward', metric = 'euclidean')
z
```
